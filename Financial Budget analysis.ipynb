```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
data = pd.read_csv(r"https://raw.githubusercontent.com/amankharwal/Website-data/master/India_budget_2021.csv")
data
```

```python
data.dropna()
```

```python
data = data.iloc[[0,8,11,14,18,23,41,42,43]]
data
```

```python
row = {"Department /Ministry":"OTHER","Fund allotted(in â‚¹crores)":592971.0800000001}
row
```

```python
data = data._append(row,ignore_index = True)
data
```

```python
data.plot.bar(x = "Department /Ministry",y = "Fund allotted(in â‚¹crores)")
```

```python
#FROM WHAT WE CAN SEE IS THAT THE LION SHARE IS TAKEN BY THE FINANCE DEPARTMENT.LETS PLOT A DONUT GRPH TO HAVE A MORE CLEAR PICTURE OF THIS .
```

```python
df = data["Fund allotted(in â‚¹crores)"]
labels = data["Department /Ministry"]
plt.figure(figsize=(7,7))
plt.pie(df, labels=labels,autopct='%1.1f%%', startangle=90, pctdistance=0.85, shadow =True)
central_circle = plt.Circle((0, 0), 0.5, color='white')
fig = plt.gcf()
fig.gca().add_artist(central_circle)
plt.rc('font', size=12)
plt.title("Distribution of The Budget", fontsize=20)
plt.show()
```

```python
#From the look of things is that the finance department has the lion share of the budget
```
